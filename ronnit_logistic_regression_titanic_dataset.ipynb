```python
# importing all the libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib
%matplotlib inline
```

```python
train_data = pd.read_csv("/content/Titanic_Data/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/content/Titanic_Data/test.csv")
test_data.head()
```

```python
y = train_data["Survived"]
features = ["Pclass", "Sex","Age", "SibSp", "Parch", "Fare"] 
#not considering features such as Name, PassengerID, Ticket, etc since they are not useful for the classification
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

#checking if the training set has null values
nan_values = X.isnull()
nan_values
for i in range(X.shape[1]):
  print('Number of missing values in feature',nan_values.columns[i],':', sum(nan_values.iloc[:,i]) )
```

Since the feature Age has a lot of missing values, it makes sense to drop that column.

```python
X_updated = X.drop(columns = ['Age'])
X_updated.head()
```

```python
X_test_updated = X_test.drop(columns = ['Age'])
X_test_updated.head()
```

Handling missing data for the test set.

```python
for i in range(X_test_updated.shape[1]):
  if sum(X_test_updated.iloc[:,i].isnull())>0:
    X_test_updated.iloc[:,i] = X_test_updated.iloc[:,i].fillna((X_test_updated.iloc[:,i]).mean())
```

```python
model = LogisticRegression()
model.fit(X_updated, y)
```

```python
predictions = model.predict(X_test_updated)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
